#### Creation of new column - Average Score

```python
df['Average Score'] = df[['math score', 'reading score',
       'writing score']].sum(axis=1)/3
```

#### First 5 Lines of dataset

```python
df.head()
```

### Test preapration and Average Score analysis

```python
sns.displot(data = df, x = "Average Score", hue = 'test preparation course',kind = "kde", aspect=2)
```

### Average Score wrt Gender

```python
sns.displot(data = df, x = "Average Score", col = 'gender', aspect=2, color='#bb98d9', kde = True)
```

### Correlation HeatMap

```python
scores = ['math score', 'reading score',
       'writing score']
sns.heatmap(df[scores].corr(), cmap="RdBu")
df[scores].corr()
```

### Maths Score wrt Gender

```python
sns.set_style('darkgrid')
sns.kdeplot(x = df['math score'], hue=df['gender'], palette="plasma")
```

```python
sns.set_style('dark')
sns.violinplot(data = df, y = "math score", x="gender", hue='race/ethnicity')
```

### Count of Race wrt to gender

```python
sns.set_style('dark')
sns.countplot(data=df, x = 'race/ethnicity', hue="gender", color="#de671d")
```
